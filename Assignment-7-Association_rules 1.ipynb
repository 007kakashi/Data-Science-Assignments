```python
import pandas as pd
import numpy as np
```

```python
# we imported the data
data=pd.read_csv('my_movies.csv')
```

```python
# we used one hot encoding method for categorical data
data1=pd.get_dummies(data)
```

```python
data1
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
# we take support 0.2 to check the how many associaction we got
freq=apriori(data1,min_support=0.2,use_colnames=True)
freq
```

```python
# now we take support 0.5 to check the how many associaction we got
freq2=apriori(data1,min_support=0.5,use_colnames=True)
freq2
```

```python
# after analyzing the both support 0.2 and 0.5 we take 0.5 support 
```

```python
asso=association_rules(freq2,metric='lift',min_threshold=1)
asso
```

```python
asso[asso['lift']>=1.5]
```

```python
asso.sort_values('lift',ascending=False)[0:20]
```

```python
# we take confidence as metrics to check the association
asso2=association_rules(freq,metric='confidence',min_threshold=1)
asso2
```

```python
asso.sort_values('confidence',ascending=False)[0:20]
```
