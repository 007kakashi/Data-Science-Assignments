```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("ToyotaCorolla.xlsx")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data=data.loc[::,["Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
```

```python
data.head()
```

```python
data.corr()
```

```python
sns.pairplot(data)
```

```python
import statsmodels.formula.api as smf
```

```python
data.columns
```

```python
data.head()
```

```python
data.corr()
```

```python
sns.regplot(data=data, x="Price",y="Age_08_04")
```

```python
model=smf.ols("Price~Age_08_04+KM+HP+Quarterly_Tax+Weight",data=data).fit()
```

```python
model.params
```

```python
model.summary()
```

```python
new_data=model.get_influence()
(c,_)=new_data.cooks_distance
```

```python
plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(data)),np.round(c,3))
```

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
influence_plot(model)
```
