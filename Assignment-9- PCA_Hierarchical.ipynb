```python
# so we do check all nessecory EDA so now we use standardization to get all values in range in -3 to +3
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled=scaler.fit_transform(data)
```

```python
scaled
```

```python
# after standardizing the value now we use PCA for feature reduction and geting maximum varience in to three components
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3,random_state=42)
pc=pca.fit_transform(scaled)
```

```python
pc
```

```python
# after that we should check the how much varience capture by 3 components 
```

```python
var=pca.explained_variance_ratio_
var
```

```python
# then we can check the pca score using 'components_' 
```

```python
pca.components_
```

```python
# now are going to use hierarchical clustering 
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# now we going to check the dendrogram to check how it is clustering using single, multiple and complete linkage 
```

```python
# first we check in single linkage
```

```python
dendrogram=sch.dendrogram(sch.linkage(pc,method='single'))
```

```python
# you can see here i use both dendrogram to check how pca helps in feature redusing and getting maximum varience in few features
```

```python
dendrogram=sch.dendrogram(sch.linkage(pc,method='single'))
```

```python
# in single linkage we it is not possible to understand so now we check with average
```

```python
dendrogram=sch.dendrogram(sch.linkage(scaled,method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(pc,method='average'))
```

```python
# now its better than single linkage now we are we are going to complete to understad which is better
```

```python
dendrogram=sch.dendrogram(sch.linkage(scaled,method='complete'))
```

```python
# in complete linkage you  will get the diffrence bettween normal and using pca clustering
```

```python
dendrogram=sch.dendrogram(sch.linkage(pc,method='complete'))
```

```python
# as you can see complete linkage is better then average and single so we use complete linakge 
```

```python
hie=AgglomerativeClustering(n_clusters=4,linkage='complete')
cluster=hie.fit(pc)
```

```python
cluster.labels_
```

```python
# now after building cluster  now this is in array from so now we are converting array into dataframe so we can do further process
```

```python
data1=pd.DataFrame(cluster.labels_)
```

```python
data1
```

```python
# after creating the dataframe of cluster feature now we are inserting into our main dataset in 1th position 
```

```python
data.insert(1,'cluster',data1)
```

```python
data
```

```python
# now we create cluster so we going to check the cluster accuracy using silhouette score
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(pc,hie.fit_predict(pc))
```
