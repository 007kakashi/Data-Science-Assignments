```python
data1=data.drop(columns=['index','month','day','DC','ISI','RH','temp'],axis=1)
```

```python
x=data1.iloc[:,:24]
y=data1.iloc[:,[-1]]
y
```

```python
# target values are imbalanced 
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
# x_train=scaler.fit_transform(x_train)
# x_test=scaler.fit_transform(x_test)
```

```python
y_train.value_counts()
```

```python
# i will use smote method to balance the imbalanced data
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
x_train,y_train=sm.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
# now we are using svm classifier with grid search cv to get best parameter
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
params={'C':[1,2,3,4,5,6,7,8,9]}
svm=SVC(random_state=0)
gv=GridSearchCV(estimator=svm,param_grid=params,cv=10)
model=gv.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
