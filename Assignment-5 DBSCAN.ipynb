```python
data1=data.rename(columns={'Unnamed: 0':'states','Murder':'murder','Assault':'assault','UrbanPop':'urbanpop','Rape':'rape'},)
```

```python
data1.head()
```

```python
x=data1.iloc[:,1:]
```

# I Used Scaler For Getting values In Scale of 0 to 1

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaled=scaler.fit_transform(x)
```

```python
scaled
```

```python
type(scaled)
```

```python
dataf=pd.DataFrame(scaled)
```

# did clustering on dataset

```python
from sklearn.cluster import DBSCAN
```

```python
cluster=DBSCAN(eps=0.2,min_samples=3)
```

```python
cluster.fit(dataf)
```

```python
labels=cluster.labels_
```

```python
labels
```

```python
pd.DataFrame(labels,columns=['cluster'])
```

# used insert to merge cluster column to merge into main dataset

```python
data1.insert(loc=0,column='cluster',value=labels)
```
